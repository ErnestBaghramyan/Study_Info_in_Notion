```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Basic Graph

```python
## Line 1
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(5,3), dpi = 200)

# Keyword Argument Notation
# plt.plot(x,y, label = "2x", color ="red", linewidth = 2, linestyle = "--", marker = ".", markersize = 10,)

# Shorthand notation
# fmt = '[color],[marker],[line]'
plt.plot(x,y, "r.--", label = "2x")


## Line 2
# Select interval we want to plot points at
x2 = np.arange(0, 4.5, 0.5)

# plt.plot(x2, x2**2, "g-.", label="X^2")

# Plot part of the graph as line
plt.plot(x2[:4], x2[:4]**2, "g-")
# Plot remainder of graph as dashed line
plt.plot(x2[3:], x2[3:]**2, "g--")

# Add a title (specify font parameters with fontdict)
plt.title("Our first graph", fontdict = {"fontname": "Courier New", "fontsize":20})

# X and Y labels
plt.xlabel("X Axis")
plt.ylabel("Y Axis")


# X, Y axis Tickmarks (scale of your graph)
plt.xticks(x) # can use the given x and y lists as ticks
plt.yticks(y)
# plt.xticks([1,2,3,4,5,6]) # can use what ever number we want to see on the graph [1,2,4,10,100], it will be resized
# plt.yticks([1,2,3,5,6,30])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig("mygraph.png", dpi=300)

# Show plot
plt.show()
```

### Bar Chart

```python
labels = ["A", "B", "C"]
values = [1,4,2]

bars = plt.bar(labels, values)

# V1 of hatching
patterns = ["/","O","*"]
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# V2 of hatching
# bars[0].set_hatch("/")
# bars[1].set_hatch("O")
# bars[2].set_hatch("*")

plt.show()
```

### Histogram

```python
bins = [40, 50, 60, 70, 80, 90, 100]

plt.hist(fifa.Overall, bins = bins, color = "#abcdef")

plt.xticks(bins)

plt.ylabel("Number of Players")
plt.xlabel("Skill Level")
plt.title("Distribution of Players Skills in FIFA 2018")

plt.show()
```

#### Pie Chart N1

```python
left = fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0]
right = fifa.loc[fifa["Preferred Foot"] == "Right"].count()[0]

labels = ["Left", "Right"]
colors = ["#abcdef", "#aabbcc"]

plt.pie([left, right], labels = labels, colors = colors, autopct = "%.2f%%")

plt.title("Foot Preference of FIFA Players")


plt.show()
```

#### Pie Chart N2

```python
fifa.Weight
```

```python
fifa.Weight = [
    int(x.strip("lbs")) if type(x)==str else x for x in fifa.Weight
]

plt.style.use("ggplot")

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ["Under 125", "125-150", "150-175", "175-200", "Over 200"]
explode = (1.5,.2,0,0, .4)

plt.pie(
    weights,
    labels = labels,
    autopct = "%.2f%%",
    pctdistance = 0.8,
    explode = explode
)

plt.title("Weight Distribution of FIFA Players (in lbs)")

plt.show()
```

```python
plt.style.use("default")

plt.figure(figsize = (9,8))

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]["Overall"]
madrid = fifa.loc[fifa.Club == "Real Madrid"]["Overall"]
revs = fifa.loc[fifa.Club == "New England Revolution"]["Overall"]
afcwimb = fifa.loc[fifa.Club == "AFC Wimbledon"]["Overall"]


labels = ["FC Barcelona", "Real Madrid", "NE Revolution", "AFC Wimbledon"]

boxes = plt.boxplot([barcelona, madrid, revs, afcwimb], labels = labels, patch_artist = True)
# We are setting "patch_artist = True" to be able to set facecolor


for box in boxes["boxes"]:
    # Set edge color
    box.set(color = "#4286f4", linewidth = 2)
    
    # Set fill color
    box.set(facecolor = "#e0e0e0")
    
plt.title("Professional Football Team Comparison")
plt.ylabel("FIFA Overall Rating")

plt.show()
```
